```python
import geopandas as gpd
import pandas as pd
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
```

```python
#Lettura file geojson
mappaEuropea = gpd.read_file("world_pre1995.geojson")

# Definire una mappa di colori graduale
cmap = plt.cm.viridis
# Normalizzare i valori delle occorrenze per la mappa di colori
norm = mcolors.Normalize(vmin=mappaEuropea['Numero'].min(), vmax=mappaEuropea['Numero'].max())

# Crea la mappa con i colori specificati e bordi neri con una linea sottile
fig, ax = plt.subplots(1, 1, figsize=(20, 10))
mappaEuropea.boundary.plot(ax=ax, linewidth=0.1, color='black')  # Linea dei bordi sottile
mappaEuropea.plot(ax=ax, color=mappaEuropea['color'], edgecolor='black')

# Aggiungere la legenda manualmente
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal', fraction=0.036, pad=0.04)
cbar.set_label('Vincitori sul podio')
ax.set_title('Nazioni con giocatori 1st fino al 1995')
ax.set_facecolor('#f0f8ff')  # Imposta il colore di sfondo a azzurrino chiaro

# Mostra la mappa
plt.show()
```

```python
#Lettura file geojson
mappaMondo = gpd.read_file("world_post1995.geojson")

# Definire una mappa di colori graduale
cmap = plt.cm.viridis
# Normalizzare i valori delle occorrenze per la mappa di colori
norm = mcolors.Normalize(vmin=mappaMondo['Numero'].min(), vmax=mappaMondo['Numero'].max())

# Crea la mappa con i colori specificati e bordi neri con una linea sottile
fig, ax = plt.subplots(1, 1, figsize=(20, 10))
mappaMondo.boundary.plot(ax=ax, linewidth=0.1, color='black')  # Linea dei bordi sottile
mappaMondo.plot(ax=ax, color=mappaMondo['color'], edgecolor='black')

# Aggiungere la legenda manualmente
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal', fraction=0.06, pad=0.04)
# Imposta dimensione del testo della legenda
cbar.ax.tick_params(labelsize=16)  # <-- dimensione dei numeri
cbar.set_label('Numero di occorenze', size=16)  # <-- dimensione delletichetta
ax.set_title('Nazioni con giocatori 1st dal 1995 ad oggi')
ax.set_facecolor('#f0f8ff')  # Imposta il colore di sfondo a azzurrino chiaro

# Mostra la mappa
plt.show()
```
