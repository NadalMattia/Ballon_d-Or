```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('dfFinalCompetition2025.csv')

firstPlaceWinners = df[df['rank'] == '1st']
secondPlaceWinners = df[df['rank'] == '2nd']
thirdPlaceWinners = df[df['rank'] == '3rd']

data = [
    firstPlaceWinners['age'].dropna(),
    secondPlaceWinners['age'].dropna(),
    thirdPlaceWinners['age'].dropna()
]

# Creazione del boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=['1° Posto', '2° Posto', '3° Posto'])
plt.title("Boxplot delle età dei vincitori del pallone d'oro")
plt.xlabel('Classifica')
plt.ylabel('Età')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

rank_mapping = {'1st': 1, '2nd': 2, '3rd': 3}
df['rank_numeric'] = df['rank'].map(rank_mapping)

# Plot the trend of ages of winners and other placements over the years
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='year', y='age', hue='rank_numeric', marker='o')
plt.title("Andamento dell'età dei vincitori del Pallone d'Oro e dei piazzamenti nel corso degli anni")
plt.xlabel('Anno')
plt.ylabel('Età')
plt.legend(title='Posizionamento', labels=['1st', '2nd', '3rd'])
plt.show()

```
