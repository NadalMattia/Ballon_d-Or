```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../Dataset/dfPlayerFinal_2025.csv')

# Filtra il dataset per includere solo i giocatori con rank "1st"
df_first_rank = df[df['rank'] == '1st']

# Crea una tabella pivot solo con i vincitori: righe = ruolo, colonne = nazionalit, valori = conteggio
pivot_first_rank = pd.pivot_table(df_first_rank, values='player', index='role', columns='nazionality', aggfunc='count', fill_value=0)

# Visualizza la heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(pivot_first_rank, annot=True, fmt='d', cmap='OrRd', linewidths=.5)

# Titolo e label
plt.title('Ruoli vs Nazionalit (solo vincitori "1st")', fontsize=16)
plt.xlabel('Nazionalit')
plt.ylabel('Ruolo giocatore')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
# Ottieni tutti i rank unici
unique_ranks = df['rank'].unique()

# Per ogni rank, crea una heatmap
for rank in unique_ranks:
    df_rank = df[df['rank'] == rank]
    pivot = pd.pivot_table(df_rank, values='player', index='role', columns='nazionality', aggfunc='count', fill_value=0)
    
    plt.figure(figsize=(15, 10))
    sns.heatmap(pivot, annot=True, fmt='d', cmap='OrRd', linewidths=.5)
    plt.title(f'Heatmap Ruoli vs Nazionalit (Rank: {rank})', fontsize=16)
    plt.xlabel('Nazionalit')
    plt.ylabel('Ruolo')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
#HeapMap unica
pivot_all_ranks = pd.pivot_table(df, values='player', index=['rank', 'role'], columns='nazionality', aggfunc='count', fill_value=0)

plt.figure(figsize=(18, 12))
sns.heatmap(pivot_all_ranks, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('Heatmap Ruoli vs Nazionalit per Tutti i Rank', fontsize=16)
plt.xlabel('Nazionalit')
plt.ylabel('Rank & Ruolo')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Crea una tabella pivot: righe = ruolo, colonne = nazionalit, valori = conteggio giocatori
pivot_all = pd.pivot_table(df, values='player', index='role', columns='nazionality', aggfunc='count', fill_value=0)

# Visualizza la heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(pivot_all, annot=True, fmt='d', cmap='OrRd', linewidths=.5)

# Titolo e label
plt.title('Frequenza Ruoli sul podio vs Nazionalit', fontsize=16)
plt.xlabel('Nazionalit')
plt.ylabel('Ruolo')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
