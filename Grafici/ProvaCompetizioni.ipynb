```python
dfFirstRank = pd.DataFrame()
dfFirstRank = dfCompetition[dfCompetition['rank'] == '1st'][['year','rank', 'player', 'team', 'nazionality']]
dfFirstRank.reset_index(drop=True, inplace=True)
dfFirstRank = dfFirstRank.drop_duplicates(subset='year')
display(dfFirstRank)
```

```python
def colonna_presente(df, colonna):
    return colonna in df.columns
```

```python
dfFirstRankCup = pd.DataFrame(dfFirstRank)
count = 0
for (_, row1), (_, row2) in zip(dfFirstRank.iterrows(), dfCoppe.iterrows()):
    team_name = row1['team']
    nazionality_name = row1['nazionality']
    for col in dfCoppe.columns:
        if row2[col] == team_name:
            if not colonna_presente(dfFirstRankCup, col):
                dfFirstRankCup[col] = None
            dfFirstRankCup.loc[count, col] = team_name
        if row2[col] == nazionality_name:
            if not colonna_presente(dfFirstRankCup, col):
                dfFirstRankCup[col] = None 
            dfFirstRankCup.loc[count, col] = nazionality_name

    count += 1

display(dfFirstRankCup)
```

```python
dfSecondtRank = pd.DataFrame()
dfSecondtRank = dfCompetition[dfCompetition['rank'] == '2nd'][['year','rank', 'player', 'team', 'nazionality']]
dfSecondtRank.reset_index(drop=True, inplace=True)
#dfSecondtRank = dfSecondtRank.drop_duplicates(subset='year')

dfSecondtRankCup = pd.DataFrame(dfSecondtRank)
count = 0
for (_, row1), (_, row2) in zip(dfSecondtRankCup.iterrows(), dfCoppe.iterrows()):
    team_name = row1['team']
    nazionality_name = row1['nazionality']
    for col in dfCoppe.columns:
        if row2[col] == team_name:
            if not colonna_presente(dfSecondtRankCup, col):
                dfSecondtRankCup[col] = None
            dfSecondtRankCup.loc[count, col] = team_name
        if row2[col] == nazionality_name:
            if not colonna_presente(dfSecondtRankCup, col):
                dfSecondtRankCup[col] = None 
            dfSecondtRankCup.loc[count, col] = nazionality_name
    count += 1
display(dfSecondtRankCup)
```

```python
dfThrerdRank = pd.DataFrame()
dfThrerdRank = dfCompetition[dfCompetition['rank'] == '3rd'][['year','rank', 'player', 'team', 'nazionality']]
dfThrerdRank.reset_index(drop=True, inplace=True)
dfThrerdRank = dfThrerdRank.drop_duplicates(subset='year')

def colonna_presente(df, colonna):
    return colonna in df.columns

dfThrerdRankCup = pd.DataFrame(dfThrerdRank)
count = 0
for (_, row1), (_, row2) in zip(dfThrerdRankCup.iterrows(), dfCoppe.iterrows()):
    team_name = row1['team']
    nazionality_name = row1['nazionality']
    for col in dfCoppe.columns:
        if row2[col] == team_name:
            if not colonna_presente(dfThrerdRankCup, col):
                dfThrerdRankCup[col] = None
            dfThrerdRankCup.loc[count, col] = team_name
        if row2[col] == nazionality_name:
            if not colonna_presente(dfThrerdRankCup, col):
                dfThrerdRankCup[col] = None  # crea la colonna se non esiste
            dfThrerdRankCup.loc[count, col] = nazionality_name
    count += 1
display(dfThrerdRankCup)
```

```python
dfTrophies = pd.DataFrame()
dfTrophies = pd.concat([dfFirstRankCup, dfSecondtRankCup, dfThrerdRankCup], ignore_index=True)
dfTrophies = dfTrophies.sort_values(by=['year', 'rank'])
dfTrophies.reset_index(drop=True, inplace=True)
dfTrophies.columns = dfTrophies.columns.str.replace('Cup', '', regex=False)
display(dfTrophies)
dfTrophies.to_csv('../Dataset/dfPlayerTrophies_2025.csv', index=False)
```

```python
for _, row in dfTrophies.iterrows():
    for col in dfTrophies.columns[5:]:
        if pd.notna(row[col]):
            if row[col] is not None:
                 dfTrophies.at[row.name, col] = col
display(dfTrophies)
dfTrophies.to_csv('../Dataset/dfPlayerTrophiesWinner_2025.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Considera solo le colonne dalla 5ª in poi (i trofei)
trophy_columns = dfTrophies.columns[5:]
display(trophy_columns)
# Sostituisci i valori non nulli con 1 per indicare trofeo vinto, altrimenti 0
df_trophies = dfTrophies.copy()
df_trophies[trophy_columns] = df_trophies[trophy_columns].notna().astype(int)

# Raggruppa per 'rank' e calcola la media per ciascun trofeo
rank_trophy_avg = df_trophies.groupby('rank')[trophy_columns].mean()

# Ordina i rank in modo coerente
rank_trophy_avg = rank_trophy_avg.loc[['1st', '2nd', '3rd']]

# Setup per il radar chart
labels = rank_trophy_avg.columns
num_vars = len(labels)

# Angoli per ciascun asse
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # chiusura del grafico

# Crea il radar plot
fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))

# Colori per i diversi rank
colors = ['#ffff00', '#00008b', '#b8860b']

for idx, (rank, row) in enumerate(rank_trophy_avg.iterrows()):
    values = row.tolist()
    values += values[:1]  # chiusura del grafico
    ax.plot(angles, values, label=rank, color=colors[idx])
    ax.fill(angles, values, color=colors[idx], alpha=0.25)

# Imposta etichette
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels, fontsize=10)
ax.set_title("Distribuzione dei trofei per Rank giocatore", size=14)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.tight_layout()
plt.show()
```
