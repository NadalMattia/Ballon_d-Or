```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Corretto l'import di matplotlib
import seaborn as sns

df = pd.read_csv('../Dataset/dfPlayerTrophiesWinner_2025.csv')
df['noTrophes'] = df.iloc[:, 5:].isna().all(axis=1).map(lambda x: 'noTrophes' if x else '')
#display(df)
#1st
dfFirst = df[df['rank'] == '1st']
dfFirstpart1 = dfFirst[dfFirst['year'] < 1995]
dfFirstpart2 = dfFirst[(dfFirst['year'] >= 1995) & (dfFirst['year'] <= 2009)]
dfFirstpart3 = dfFirst[(dfFirst['year'] >= 2010) & (dfFirst['year'] <= 2015)]
dfFirstpart4 = dfFirst[dfFirst['year'] > 2015]
#2nd
dfSecond = df[df['rank'] == '2nd']
dfSecondpart1 = dfSecond[dfSecond['year'] < 1995]
dfSecondpart2 = dfSecond[(dfSecond['year'] >= 1995) & (dfSecond['year'] <= 2009)]
dfSecondpart3 = dfSecond[(dfSecond['year'] >= 2010) & (dfSecond['year'] <= 2015)]
dfSecondpart4 = dfSecond[dfSecond['year'] > 2015]
#3rd
dfThrerd = df[df['rank'] == '3rd']
dfThrerdpart1 = dfThrerd[dfThrerd['year'] < 1995]
dfThrerdpart2 = dfThrerd[(dfThrerd['year'] >= 1995) & (dfThrerd['year'] <= 2009)]
dfThrerdpart3 = dfThrerd[(dfThrerd['year'] >= 2010) & (dfThrerd['year'] <= 2015)]
dfThrerdpart4 = dfThrerd[dfThrerd['year'] > 2015]
```

```python
df_95 = df[df['year'] < 1995]
df_95_09 = df[(df['year'] >= 1995) & (df['year'] <= 2009)]
df_10_15 = df[(df['year'] >= 2010) & (df['year'] <= 2015)]
df_16 = df[df['year'] > 2015]
```

```python
def compute_counts(df, label):
    counts = {}
    for col in df.columns[5:]:
        counts[col] = (df[col] == col).sum()
    return pd.DataFrame({
        'Competizione': list(counts.keys()),
        'Count': list(counts.values()),
        'Periodo': label
    })

counts_pre95 = compute_counts(dfFirstpart1, 'pre-1995')
counts_95_09 = compute_counts(dfFirstpart2, '19952009')
counts_10_15 = compute_counts(dfFirstpart3, '20102015')
counts_16_24 = compute_counts(dfFirstpart4, 'post-2016')

df_all = pd.concat([counts_pre95, counts_95_09, counts_10_15, counts_16_24], ignore_index=True)

# Crea il grafico a barre con somma invece della media
plt.figure(figsize=(12, 6))

sns.barplot(
    data=df_all,
    x='Competizione',
    y='Count',
    hue='Periodo',
    palette='Set2',
    estimator=sum,  # Usa la somma invece della media
    ci=None  # Disabilita intervallo di confidenza (per non avere valori decimali)
)

# Aggiungi etichette per l'asse y interi
plt.yticks(range(0, int(max(df_all['Count'])) + 1))

# Abbellisci il grafico
plt.xticks(rotation=90)
plt.title('Competizioni da Rank 1nd raggruppate per annate')
plt.ylabel('Numero di volte al 1 posto')
plt.xlabel('Competizione')
plt.tight_layout()
plt.show()
```

```python
#Calcolo occorrenze

import pandas as pd
import matplotlib.pyplot as plt

# Funzione per calcolare i conteggi
def compute_counts(df, label, rank):
    counts = {}
    for col in df.columns[5:]:
        counts[col] = (df[col] == col).sum()
    return pd.DataFrame({
        'Competizione': list(counts.keys()),
        'Count': list(counts.values()),
        'Periodo': label,
        'Rank': rank
    })

# Calcola i conteggi per ogni periodo e rank (1st, 2nd, 3rd)
counts_pre95_first = compute_counts(dfFirstpart1, 'pre-1995', '1st')
counts_95_09_first = compute_counts(dfFirstpart2, '19952009', '1st')
counts_10_15_first = compute_counts(dfFirstpart3, '20102015', '1st')
counts_16_24_first = compute_counts(dfFirstpart4, 'post-2016', '1st')

counts_pre95_second = compute_counts(dfSecondpart1, 'pre-1995', '2nd')
counts_95_09_second = compute_counts(dfSecondpart2, '19952009', '2nd')
counts_10_15_second = compute_counts(dfSecondpart3, '20102015', '2nd')
counts_16_24_second = compute_counts(dfSecondpart4, 'post-2016', '2nd')

counts_pre95_third = compute_counts(dfThrerdpart1, 'pre-1995', '3rd')
counts_95_09_third = compute_counts(dfThrerdpart2, '19952009', '3rd')
counts_10_15_third = compute_counts(dfThrerdpart3, '20102015', '3rd')
counts_16_24_third = compute_counts(dfThrerdpart4, 'post-2016', '3rd')

# Unisci tutti i dati
df_all = pd.concat([
    counts_pre95_first, counts_95_09_first, counts_10_15_first, counts_16_24_first,
    counts_pre95_second, counts_95_09_second, counts_10_15_second, counts_16_24_second,
    counts_pre95_third, counts_95_09_third, counts_10_15_third, counts_16_24_third
], ignore_index=True)

# Rimuoviamo eventuali NaN nel caso ci siano valori nulli
df_all = df_all.dropna(subset=['Competizione', 'Count'])
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Palette personalizzata per i Rank
custom_palette = {
    '1st': '#FFD700',   # Oro
    '2nd': '#C0C0C0',   # Argento
    '3rd': '#CD7F32'    # Bronzo
}

# Titoli leggibili per i periodi
period_titles = {
    'pre-1995': 'Prima del 1995',
    '19952009': '19952009',
    '20102015': '20102015',
    'post-2016': 'Dal 2016 in poi'
}

# Crea un grafico per ciascun periodo
periods = ['pre-1995', '19952009', '20102015', 'post-2016']
period_titles = {
    'pre-1995': 'Prima del 1995',
    '19952009': '19952009',
    '20102015': '20102015',
    'post-2016': 'Dal 2016 in poi'
}

for period in periods:
    # Filtra solo i dati per il periodo corrente
    df_period = df_all[df_all['Periodo'] == period]

    plt.figure(figsize=(12, 6))
    sns.barplot(
        data=df_period,
        x='Competizione',
        y='Count',
        hue='Rank',  # solo Rank viene usato per colorare
        palette=custom_palette
    )

    # Mostra solo valori interi sull'asse Y
    max_count = int(df_period['Count'].max())
    plt.yticks(range(0, max_count + 2))  # +2 per margine finale
    
    plt.title(f'Posizionamenti (1, 2, 3) per Competizione - {period_titles[period]}')
    plt.ylabel('Numero di volte sul podio')
    plt.xlabel('Competizione')
    plt.xticks(rotation=90)
    plt.legend(title='Rank')  # Nessun riferimento al periodo
    plt.tight_layout()
    plt.show()
```

```python
#Grafici subplot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Palette personalizzata per i Rank
custom_palette = {
    '1st': '#FFD700',   # Oro
    '2nd': '#C0C0C0',   # Argento
    '3rd': '#CD7F32'    # Bronzo
}

# Titoli leggibili per i periodi
period_titles = {
    'pre-1995': 'Prima del 1995',
    '19952009': '19952009',
    '20102015': '20102015',
    'post-2016': 'Dal 2016 in poi'
}

# Crea una figura con 2 righe e 2 colonne di grafici
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Lista dei periodi per ciclo
periods = ['pre-1995', '19952009', '20102015', 'post-2016']

# Ciclo per tracciare ogni grafico
for i, period in enumerate(periods):
    ax = axes[i//2, i%2]  # Seleziona il sotto-grafico appropriato
    # Filtra solo i dati per il periodo corrente
    df_period = df_all[df_all['Periodo'] == period]

    sns.barplot(
        data=df_period,
        x='Competizione',
        y='Count',
        hue='Rank',  # solo Rank viene usato per colorare
        palette=custom_palette,
        ax=ax  # Passa l'asse specifico
    )

    # Mostra solo valori interi sull'asse Y
    max_count = int(df_period['Count'].max())
    ax.set_yticks(range(0, max_count + 2))  # +2 per margine finale
    
    ax.set_title(f'Posizionamenti (1, 2, 3) per Competizione - {period_titles[period]}')
    ax.set_ylabel('Numero di volte sul podio')
    ax.set_xlabel('Competizione')
    ax.tick_params(axis='x', rotation=90)  # Ruota le etichette dell'asse x
    ax.legend(title='Rank')  # Nessun riferimento al periodo

# Ottimizza la disposizione dei grafici
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  # Assicurati che dfFirstpart1 sia definito

# Calcolo dei conteggi
counts = {}
for col in dfFirstpart1.columns[5:]:
    counts[col] = (dfFirstpart1[col] == col).sum()

labels = list(counts.keys())
values = list(counts.values())

# Evita problemi di log(0): sostituisci 0 con 1 (o un valore minimo positivo)
values_log = [np.log10(v if v > 0 else 1) for v in values]

# Chiudi il loop
values_log += values_log[:1]

# Setup angoli
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Disegna
fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))
ax.plot(angles, values_log, color='green', linewidth=2)
# ax.fill(angles, values_log, color='green', alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, size=8)

# Tick radiali logaritmici (in scala log10, ad esempio da 0 a log10(max))
max_log = max(values_log)
yticks = np.linspace(0, max_log, num=17)
ax.set_yticks(yticks)
ax.set_yticklabels([f"{10**ytick:.0f}" for ytick in yticks])  # Mostra etichette come valori originali

# Titolo
plt.title('Competizioni da Rank 1st prima del 1995 (Scala Log)', size=15, y=1.1)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Supponiamo che dfFirstpart1 sia gi caricato e abbia una colonna 'Anno'
# Crea colonna 'Period' per dividere prima/dopo il 1995
dfFirstpart1['Period'] = dfFirstpart1['year'].apply(lambda x: 'Before 1995' if x < 1995 else 'After 1995')

# Trasforma in formato long, escludendo le prime 5 colonne (presumibilmente metadata) e 'Anno' e 'Period'
df_long = pd.melt(
    dfFirstpart1,
    id_vars=['Period'],
    value_vars=dfFirstpart1.columns[5:-1],  # usa solo le colonne di competizione
    var_name='Competition',
    value_name='Value'
)

# Filtra solo i casi in cui il valore corrisponde al nome della competizione (cio, 1st place)
df_long_filtered = df_long[df_long['Value'] == df_long['Competition']]

# Conta quanti 1st places per competizione e periodo
counts = df_long_filtered.groupby(['Period', 'Competition']).size().reset_index(name='Count')

# Crea il FacetGrid con un barplot per periodo
g = sns.FacetGrid(counts, col="Period", col_wrap=2, height=5, sharey=False)
g.map_dataframe(sns.barplot, x='Competition', y='Count', palette='Greens_d')

# Migliora la leggibilit delle etichette
for ax in g.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

# Titolo complessivo
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle('1st Place per Competizione (prima e dopo il 1995)', fontsize=16)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crea colonna 'Period'
dfFirstpart1['Period'] = dfFirstpart1['year'].apply(lambda x: 'Before 1995' if x < 1995 else 'After 1995')

# Seleziona colonne delle competizioni
comp_cols = dfFirstpart1.columns[5:-1]  # exclude 'Anno' and 'Period'

# Prepara una figura con un radar plot per ogni periodo
periods = dfFirstpart1['Period'].unique()
num_plots = len(periods)

fig, axes = plt.subplots(1, num_plots, figsize=(6*num_plots, 6), subplot_kw=dict(polar=True))

# Se  un solo plot, lo mettiamo in una lista
if num_plots == 1:
    axes = [axes]

for i, period in enumerate(periods):
    subset = dfFirstpart1[dfFirstpart1['Period'] == period]
    counts = {}
    for col in comp_cols:
        counts[col] = (subset[col] == col).sum()

    labels = list(counts.keys())
    values = list(counts.values())

    # Setup angoli
    num_vars = len(labels)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    values += values[:1]  # chiude il grafico
    angles += angles[:1]

    ax = axes[i]
    ax.plot(angles, values, color='green', linewidth=2)
    ax.fill(angles, values, color='green', alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, size=8)
    ax.set_title(f'1st Place - {period}', size=14, y=1.1)

plt.suptitle('Radar Plot 1st Place per Competizione', size=18, y=1.05)
plt.tight_layout()
plt.show()
```
