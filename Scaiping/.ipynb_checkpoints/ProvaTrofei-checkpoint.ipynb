```python
#Champions

import requests
from bs4 import BeautifulSoup

# URL della pagina Wikipedia
url = "https://it.wikipedia.org/wiki/UEFA_Champions_League"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

colonne_divs = soup.find_all("div", class_="colonne")

# Lista dei dizionari anno-squadra
championsWinner = []

for div in colonne_divs:
    for ul in div.find_all("ul"):
        for li in ul.find_all("li"):
            a_tags = li.find_all("a", href=True)
            if len(a_tags) >= 2:
                href1 = a_tags[0]['href']
                year = href1[-4:]

                href2 = a_tags[3].get_text(strip=True)
                team = href2

                championsWinner.append({
                    "year": year,
                    "team": team
                })

print(championsWinner)
```

```python
#SerieA

url = "https://it.wikipedia.org/wiki/Serie_A"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

colonne_divs = soup.find_all("div", class_="colonne_strette")

serieA_Winner = []

for div in colonne_divs:
    for ul in div.find_all("ul"):
        for li in ul.find_all("li"):
            a_tags = li.find_all("a", href=True)
            if len(a_tags) >= 2:
                href1 = a_tags[0].get_text(strip=True)
                if (len(href1) > 4):
                    href1 = href1[-4:]
                year = href1

                href2 = a_tags[1]['title']
                team = href2

                serieA_Winner.append({
                    "year": year,
                    "team": team
                })

serieA_Winner = [voce for voce in serieA_Winner if voce['year'].isdigit() and int(voce['year']) > 1955]
print(serieA_Winner)
```

```python
#Eredivisie (campionato Olandese)

url = "https://it.wikipedia.org/wiki/Eredivisie"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

colonne_divs = soup.find_all("div", class_="colonne_strette")

# Lista dei dizionari anno-squadra
eredivisie_Winner = []

for div in colonne_divs:
    for ul in div.find_all("ul"):
        for li in ul.find_all("li"):
            a_tags = li.find_all("a", href=True)
            href1 = a_tags[0]['href']
            if (len(href1) > 4):
                href1 = href1[-4:]
            year = href1

            b_tag = li.find('b')
            if b_tag:
                a_tags = b_tag.find_all('a')
                if a_tags:
                    team = a_tags[-1].text.strip()

            eredivisie_Winner.append({
                "year": year,
                "team": team
            })

eredivisie_Winner = [voce for voce in eredivisie_Winner if voce['year'].isdigit() and int(voce['year']) > 1955]
print(eredivisie_Winner)
```

```python
#Superliga (Danimarca)

url = "https://it.wikipedia.org/wiki/Superligaen"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

colonne_divs = soup.find_all("div", class_="colonne_strette")

# Lista dei dizionari anno-squadra
superliga_Winner = []

for div in colonne_divs:
    for ul in div.find_all("ul"):
        for li in ul.find_all("li"):
            a_tags = li.find_all("a", href=True)
            href1 = a_tags[0]['href']
            if (len(href1) > 4):
                href1 = href1[-4:]
            year = href1

            b_tag = li.find('b')
            if b_tag:
                a_tags = b_tag.find_all('a')
                if a_tags:
                    team = a_tags[-1].text.strip()
            else:
                team = a_tags[-1].text.strip()

            superliga_Winner.append({
                "year": year,
                "team": team
            })

superliga_Winner = [voce for voce in superliga_Winner if voce['year'].isdigit() and int(voce['year']) > 1955]
print(superliga_Winner)
```

```python
#Scottish Premiership (Scozia)

url = "https://it.wikipedia.org/wiki/Scottish_Premiership"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

colonne_divs = soup.find_all("div", class_="colonne_strette")

# Lista dei dizionari anno-squadra
scottishPremiership_Winner = []

for div in colonne_divs:
    for ul in div.find_all("ul"):
        for li in ul.find_all("li"):
            a_tags = li.find_all("a", href=True)
            if not a_tags:
                continue
            href1 = a_tags[0]['href']
            if (len(href1) > 4):
                href1 = href1[-4:]
            year = href1

            b_tag = li.find('b')
            if b_tag:
                a_tags2 = b_tag.find_all('a')
                if a_tags:
                    team = a_tags2[-1].text.strip()
            else:
                team = a_tags[2].text.strip()

            scottishPremiership_Winner.append({
                "year": year,
                "team": team
            })

scottishPremiership_Winner = [voce for voce in scottishPremiership_Winner if voce['year'].isdigit() and int(voce['year']) > 1955]
print(scottishPremiership_Winner)
```
