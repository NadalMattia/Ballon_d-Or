```python
import pandas as pd
df = pd.read_csv('dfFinalCompetition2025.csv')
df_Champions_EuropaLeague = pd.read_csv('datasetChampionEuropaLeague.csv')
df_CompetizioniNazionali = pd.read_csv('datasetCompetizioniNazioneali.csv')
df_CompetizioniPerNazione = pd.read_csv('datasetCompetizioniPerNazione.csv')

display(df)
display(df_Champions_EuropaLeague)
display(df_CompetizioniNazionali)
display(df_CompetizioniPerNazione)
```

```python
df_competitions = pd.DataFrame(df_Champions_EuropaLeague)
df_competitions = pd.merge(df_competitions, df_CompetizioniNazionali, on='year', how='outer')
df_competitions = df_competitions.sort_values(by='year').reset_index(drop=True)

display(df_competitions)
```

```python
df_Unione =  pd.DataFrame(df_Champions_EuropaLeague)
df_Unione = pd.merge(df_Unione, df_CompetizioniNazionali, on='year', how='outer')
df_Unione = pd.merge(df_Unione, df_CompetizioniPerNazione, on='year', how='outer')
df_Unione = df_Unione.sort_values(by='year').reset_index(drop=True)

display(df_Unione)
df_Unione.to_csv('dfUnioneCoppe.csv', index=False)
```
